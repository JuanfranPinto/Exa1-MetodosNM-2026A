```python
import numpy as np
from scipy import optimize

# Definición del polinomio
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

##metodo de la secante
def encontrar_raices():
    raices = []

    
    puntos = np.linspace(-10, 10, 40)

    for i in range(len(puntos) - 1):
        x0 = puntos[i]
        x1 = puntos[i + 1]

        try:
            raiz = optimize.newton(
                f,
                x0=x0,
                x1=x1,      
                tol=1e-10,
                maxiter=100
            )

            # Verificar que sea real
            if np.isreal(raiz):
                raiz = float(np.real(raiz))

                # Evitar raíces repetidas
                if all(abs(raiz - r) > 1e-5 for r in raices):
                    raices.append(raiz)

        except:
            pass

    # Ordenar las raíces
    raices.sort()

    # Completar hasta 6 valores con NaN
    while len(raices) < 6:
        raices.append(np.nan)

    return raices[:6]

# Obtener raíces
resultado = encontrar_raices()

# mostras decirmales
print("raices:\n")

for i, r in enumerate(resultado, start=1):
    if np.isnan(r):
        print(f"x{i} = NaN")
    else:
        print(f"x{i} = {r:.10f}")
```
