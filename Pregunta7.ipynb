```python
import numpy as np
import math

def redondeo(num, cifras):
    if num == 0:
        return 0
    magnitud = math.floor(math.log10(abs(num)))
    decimales = cifras - 1 - magnitud
    return round(num, decimales)


def metodo_logaritmo(x):
    temp = x

    # 5 raíces cuadradas consecutivas
    for _ in range(5):
        temp = math.sqrt(temp)
        temp = redondeo(temp, 6)

    r = temp - 1
    alpha = 1 / math.log(10)

    log_aprox = 32 * alpha * r
    log_aprox = redondeo(log_aprox, 6)

    return log_aprox, r


def calculo_error(log_calcu, log_real):
    error = abs(log_real - log_calcu) / abs(log_real)
    return redondeo(error, 6)

#Ejemplo con x = 2
x = 2
logaritmo, rmas1 = metodo_logaritmo(x)

print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 3
x= 3
logaritmo, rmas1 = metodo_logaritmo(x)
print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 4
x= 4
logaritmo, rmas1 = metodo_logaritmo(x)
print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 5
x = 5
logaritmo, rmas1 = metodo_logaritmo(x)

print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 6
x= 6
logaritmo, rmas1 = metodo_logaritmo(x)
print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 7
x= 7
logaritmo, rmas1 = metodo_logaritmo(x)
print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 8
x= 8
logaritmo, rmas1 = metodo_logaritmo(x)
print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 9
x= 9
logaritmo, rmas1 = metodo_logaritmo(x)
print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))

# Ejemplo con x = 10
x = 10
logaritmo, rmas1 = metodo_logaritmo(x)
print('logaritmo base 10 calculado:', logaritmo, 'para:', x)
print('logaritmo real:', np.log10(x))
print('error relativo:', calculo_error(logaritmo, np.log10(x)))
```
